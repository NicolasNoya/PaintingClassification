# EXPERIMENT 2.1: Training the 2 Branch RESNET50 
# with Augmentation

In this experiment, we train a custom Two-Branch ResNet50 architecture using an augmented version of our dataset. After training, we evaluate the model’s performance and visualize correctly and incorrectly classified examples for both Abstract and Figurative classes.

```python
from interface import Interface, ModelsName
import torch
```

```python
interface = Interface(epochs=50, batch_size=32,logging_interval=10,
            n_transforms_augmented=2,num_workers=16,model=ModelsName.TWO_RESNET,
            save_model_path='weigths/2branchRNN/20250619_augmented/',data_path='../new_dataset/',
            augmentation = True, noise = False)
```

```python
interface.rnn_train()
```

```python
running_loss, metric_dict, [correct_abstract,wrong_abstract,correct_figurative,wrong_figurative] = interface.test()
```

```python
metric_dict
```

```python
import matplotlib.pyplot as plt

def show_images(img_list, title):
    fig, axes = plt.subplots(1, len(img_list), figsize=(15, 3))
    fig.suptitle(title)
    for i, (img, label, pred, prob) in enumerate(img_list):
        axes[i].imshow(img)
        axes[i].set_title(f"Label: {label}\nPred: {pred} ({prob:.2f})")
        axes[i].axis('off')
    plt.tight_layout()
    plt.show()


show_images(correct_abstract, "Correct Abstract")
show_images(wrong_abstract, "Wrong Abstract")
show_images(correct_figurative, "Correct Figurative")
show_images(wrong_figurative, "Wrong Figurative")
```
